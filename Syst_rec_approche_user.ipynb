<h1>Système de recommandations de films avec des méthodes de machine learning : approche 'USER'</h1>

```python
import pandas as pd
import numpy as np
```

<h5>----------------------------dataset ratings-----------------------------------</h5>

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings.info()
```

<h5>----------------------------dataset movies-----------------------------------</h5>

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

<h5>----------------------------merge datasets-----------------------------------</h5>

```python
df = pd.merge(ratings, movies,on='movieId', how='inner')
df.head()
```

<h5>---------------------------- Matrice User/movie/rating -----------------------------------</h5>

```python
matrice=df.pivot_table(index ='userId', columns ='title', values ='rating')
matrice
```

```python
matrice=matrice.dropna(thresh=50, axis=1)
matrice
```

<h5>------------------------------- Matrice normalisée -----------------------------------------</h5>

```python
matrice_norm=matrice.subtract(matrice.mean(axis=1), axis='rows')
matrice_norm
```

<h5>------------------------------- Matrice de similarité entre utilisateurs-----------------------------------------</h5>

```python
matrice_similarity=matrice_norm.T.corr()
matrice_similarity
```

<h5>------------------------------- PREDICTION / RECOMMANDATION POUR UN UTILISATEUR --------------------------------------</h5>

```python
def matrice_norm_cleaned(user):
    
    similar_users=matrice_similarity[[user]].rename(columns={user:'similarity'})

    similar_users=similar_users[similar_users.similarity>0].sort_values(by='similarity', ascending=False)

    similar_users.drop(index=user, inplace=True)
    

    unrated_movies_user=matrice_norm.loc[user].reset_index()
    unrated_movies_user=unrated_movies_user[unrated_movies_user[user].isna()]['title']
    
    similar_users2=similar_users.reset_index()
    matrice_similarity_recommendation=matrice_norm[matrice_norm.index.isin(similar_users2['userId'])]
    matrice_similarity_recommendation=matrice_similarity_recommendation[unrated_movies_user]
    
    
    movie_to_recommend=matrice_similarity_recommendation.dropna(axis=0, how='all')
    
    return similar_users, movie_to_recommend
    
```

```python
def get_recommendation(user):
    
    new_recommendation=[]
    
    similar_users, movie_to_recommend = matrice_norm_cleaned(user)

    List_recommendation=pd.DataFrame(columns=["Title", "Prediction - Rating"])
    
    mean_value=matrice.loc[user,:].mean()

    for movie in movie_to_recommend.columns:

        rating_norm=movie_to_recommend[[movie]].dropna().reset_index()

        rating_similarity=pd.merge(rating_norm, similar_users, on='userId')

        rating_similarity['Rating']=rating_similarity.apply(lambda x: (x[movie]*x['similarity']), axis=1)    
        prediction=(rating_similarity['Rating'].sum())/(rating_similarity['similarity'].sum())
        
        prediction+=mean_value

        new_recommendation=[movie, prediction]
        List_recommendation.loc[len(List_recommendation)]=new_recommendation
        
    List_recommendation=List_recommendation.sort_values(by='Prediction - Rating', ascending=False)
    return List_recommendation.head(10)
```

```python
get_recommendation(1)
```
