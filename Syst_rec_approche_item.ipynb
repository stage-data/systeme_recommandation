<h1>Système de recommandations de films avec des méthodes de machine learning : approche 'ITEM'</h1>

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

<h5>----------------------------dataset ratings-----------------------------------</h5>

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
ratings.info()
```

<h5>----------------------------dataset movies-----------------------------------</h5>

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
movies.info()
```

<h5>----------------------------merge datasets-----------------------------------</h5>

```python
df = pd.merge(ratings, movies)
df.head()
```

<h5>---------------------------- Matrice Movie/user/rating -----------------------------------</h5>

```python
matrice=df.pivot_table(index ='title', columns ='userId', values ='rating')
matrice
```

```python
matrice=matrice.dropna(thresh=50, axis=0)
matrice
```

<h5>------------------------------- Matrice normalisée -----------------------------------------</h5>

```python
matrice_norm=matrice.subtract(matrice.mean(axis=1), axis='rows')
matrice_norm
```

<h5>------------------------------- Matrice de similarité -----------------------------------------</h5>

```python
matrice_similarity=matrice_norm.fillna(0)
matrice_similarity=cosine_similarity(matrice_similarity)
matrice_similarity=pd.DataFrame(matrice_similarity, index=matrice_norm.index, columns=matrice_norm.index)
matrice_similarity
```

<h5>------------------------------- PREDICTION / RECOMMANDATION POUR UN UTILISATEUR --------------------------------------</h5>

```python
def get_recommendation(user):
    
    new_recommendation=[]


    List_recommendation=pd.DataFrame(columns=["Title", "Prediction - Rating"])

    rated_movies=pd.DataFrame(matrice_norm[user][~matrice_norm[user].isna()].reset_index())
    rated_movies=rated_movies.rename(columns={user:'rating'})

    unrated_movies=pd.DataFrame(matrice_norm[user][matrice_norm[user].isna()].reset_index())
    unrated_movies=unrated_movies['title'].tolist()

    
    for movie in unrated_movies:   

        similar_movies=matrice_similarity[movie][matrice_similarity[movie]>0]
        similar_movies=similar_movies.filter(items=rated_movies['title'].tolist(), axis=0).to_frame().reset_index()
        similar_movies=similar_movies.rename(columns={movie:'similarity'})

        rating_similarity=pd.merge(rated_movies, similar_movies, on='title').reset_index()#.sort_values('similarity', ascending=False)[:5]

        rating_similarity['prediction']=rating_similarity.apply(lambda x: (x['rating']*x['similarity']), axis=1)    
        prediction=(rating_similarity['prediction'].sum())/(rating_similarity['similarity'].sum())

        mean_value=matrice.loc[movie,:].mean()
        prediction+=mean_value

        new_recommendation=[movie, prediction]
        List_recommendation.loc[len(List_recommendation)]=new_recommendation

    List_recommendation=List_recommendation.sort_values(by='Prediction - Rating', ascending=False)
    return List_recommendation.head(10)
```

```python
get_recommendation(1)
```
